# 3.2.2 Recognising data that violates normality assumptions

Follow the demonstration to learn how to recognise data that violates normality assumptions. In this demonstration, you will learn how to:
- Apply graphical methods.
- Apply statistical tests.
- Interpret the output of these tests.

Consider the following scenario: You were tasked to determine whether a random data set is normally distributed or not. The hypotheses you will test with $\alpha=0.05$, are:
- $H_0$: The data is normally distributed.
- $H_a$: The data is not normally distributed.

> Please note that we will create a random data set. Therefore, your output might slightly differ from the video.

## a. Graphical methods (visual checks)
We will explore two methods: histogram and a Q-Q plot.

### Histogram

```python
# Import the necessary libraries.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a data set consisting of 300 values.
np.random.seed(142)

data = np.random.uniform(low=1,
                         high=10,
                         size=300)
```

```python
# Create a histogram.
sns.histplot(data,
             bins=10);

# Specify labels and a title.
plt.xlabel("Sample data", size=15);
plt.ylabel("Density", size=15);
plt.title("Histogram", size=20);

# View the plot.
plt.show()
```

### Q-Q plot

```python
# Plot a Q-Q plot.
stats.probplot(data,
               dist="norm",
               plot=plt)

# View the plot.
plt.show()
```

## b) Statistical test (Shapiro–Wilk test)

```python
# Import the necessary libraries.
from scipy.stats import shapiro
```

```python
# Perform the Shapiro–Wilk test.
stat, p = shapiro(data)
print("Shapiro-Wilk statistic:", stat)
print("Shapiro-Wilk p-value:", p)
```

```python
# Interpret.
alpha = 0.05
if p > alpha:
    print('Sample looks normally distributed  (fail to reject H0)')
else:
    print('Sample does not look normally distributed (reject H0)')
```
