# Demonstration 2.2.1 Measuring correlation


Follow the demonstration to learn how to measure correlation with the Pearson and Spearman methods. In this video, you will learn how to:
- Perform and interpet Pearson correlation.
- Perform and interpret Spearman correlation.


## a. Create and explore the data set

```python
# Import the required libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create an expanded data set with more data points.
url = "https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/refs/heads/main/C1W2/weight-height.csv"

# Convert to a Pandas DataFrame.
df = pd.read_csv(url)
```

```python
# View the DataFrame.
print(df.shape)
df
```

```python
print(df.dtypes)
```

```python
# Exploratory data analysis (EDA).
# Descriptive Statistics.
df.describe(include = "all")
```

```python
# Visualise possible correlation.
plt.figure(figsize=(10, 8))

# Create a scatterplot.
plt.scatter(df['Height'],
            df['Weight'],
            c='b')

plt.title('Scatterplot')
plt.xlabel('Height')
plt.ylabel('Weight')

# Show the plot.
plt.show()
```

```python
# Visualise possible correlation.
plt.figure(figsize=(10, 8))

# Create a scatterplot.
plt.hist(df['Height'],
         color = 'g',
         alpha = 0.7,
         label = 'Height',
         bins = 10)

plt.title('Histogram of height')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.legend()

# Show the plot.
plt.show()
```

```python
# Visualise possible correlation.
plt.figure(figsize=(10, 8))

plt.hist(df['Weight'],
         color = 'r',
         alpha = 0.7,
         label = 'Weight',
         bins = 10)
plt.title('Histogram of weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.legend()

# Show the plot.
plt.show()
```

## b. Pearson correlation

```python
# Import the required library.
from scipy.stats import pearsonr

# Calculate Pearson correlation.
pearson_corr, pearson_p_value = pearsonr(df['Height'],
                                         df['Weight'])

# View the output.
print("\nPearson correlation coefficient:", pearson_corr)
print("Pearson p-value:", pearson_p_value)
```

```python
# Visualise a heatmap to display the correlation matrix.
plt.figure(figsize=(10, 8))
corr_matrix_pearson = df.corr(method='pearson')

sns.heatmap(corr_matrix_pearson,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            annot_kws={"fontsize": 16, "fontweight": "bold"})

plt.title('Pearson correlation')
plt.show()
```

## c. Create and explore the data set

```python
# Import the data set.
url ='https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/thurber.csv'

# Convert to a Pandas DataFrame.
df2 = pd.read_csv(url)
```

```python
# View the DataFrame.
print(df2.shape)
df2
```

```python
print(df2.dtypes)
```

```python
# Descriptive Statistics.
df2.describe(include = "all")
```

```python
# Visualise possible correlation.
plt.figure(figsize=(10, 8))

# Create a scatterplot.
plt.scatter(df2['x'],
            df2['y'],
            c='b')

plt.title('Scatterplot')
plt.xlabel('Predictor variable')
plt.ylabel('Response variable')

# Show the plot.
plt.show()
```

## d. Spearman correlation

```python
# Import the required library.
from scipy.stats import spearmanr

# Calculate the Spearman correlation.
spearman_corr, spearman_p_value = spearmanr(df2['x'],
                                            df2['y'])

# View the output.
print("\nSpearman correlation coefficient:", spearman_corr)
print("Spearman p-value:", spearman_p_value)
```

```python
# Visualise using a correlation matrix heatmap.
plt.figure(figsize=(10, 8))
corr_matrix_spearman = df2.corr(method='spearman')

sns.heatmap(corr_matrix_spearman,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            annot_kws={"fontsize": 16, "fontweight": "bold"})

plt.title('Spearman correlation')

# Show the plot.
plt.show()
```
